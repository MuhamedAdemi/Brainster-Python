## Problem 1
Let $f(x) = x^4 - x^3$.
* Find $f'(x)$ and hence find the critical points of the function.
* Using $f''(x)$, establish the nature of the critical points (i.e. if they are min, max, or neither).
* Calculate any local max/min values the function may have.

```python

# define variables and functions:
x = sp.symbols('x')  # Define x as a symbolic variable
f = x**4 - x**3      # Define the function f(x) = x^4 - x^3

#  Find the first derivative f'(x)
f_prime = sp.diff(f, x)  # Derivative of f with respect to x
print("First derivative, f'(x):", f_prime)

# Find the critical points by solving f'(x) = 0
critical_points = sp.solve(f_prime, x)
print("Critical points:", critical_points)

# Find the second derivative f''(x)
f_double_prime = sp.diff(f_prime, x)  # Second derivative
print("Second derivative, f''(x):", f_double_prime)
# Test the critical points using second derivative
#  Test the nature of the critical points
for point in critical_points:
    second_derivative_value = f_double_prime.subs(x, point)
    if second_derivative_value > 0:
        nature = "Local Minimum"
    elif second_derivative_value < 0:
        nature = "Local Maximum"
    else:
        nature = "Inconclusive"
    print(f"At x = {point}, f''(x) = {second_derivative_value}, Nature: {nature}")

# 5. Calculate function values at critical points
for point in critical_points:
    function_value = f.subs(x, point)
    print(f"f({point}) = {function_value}")

# Just to see if we are ok with the conclusion, graph:
# 6. Plot the function to confirm
x_vals = np.linspace(-0.5, 1.5, 500)  # Values for x in the range of interest
f_vals = [f.subs(x, val) for val in x_vals]  # Evaluate f(x) for all x values

plt.figure(figsize=(8, 6))
plt.plot(x_vals, f_vals, label='f(x) = x^4 - x^3', color='blue')
plt.scatter(critical_points, [f.subs(x, cp) for cp in critical_points], color='red', label='Critical Points')
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.axvline(0, color='black', linewidth=0.8, linestyle='--')
plt.title('Graph of f(x) = x^4 - x^3')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid()
plt.show()
```

## Problem 3
A manufacturer has determined that the cost $C$ of operating a factory is given by
\begin{equation}C(x)=0.5x^2+10x+7200,\end{equation}
where $x$ is the number of units produced.
* At what level of production will the **average cost** per unit be minimized? (Note: the average cost is the ratio of the cost of producing $x$ units and $x$, the number of produced units).
* What is the minimal average cost per unit produced?

```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt

# ---------------------------------------------------------------
# Define variables and functions
# ---------------------------------------------------------------
x = sp.symbols('x')

# Example cost function (C(x))
C = 0.01*x**3 - 0.6*x**2 + 13*x + 500

# Average cost function A(x) = C(x) / x
A = C / x

# ---------------------------------------------------------------
# Find critical points of A(x)
# ---------------------------------------------------------------
A_prime = sp.diff(A, x)  # First derivative

# Solve A'(x) = 0 to find critical points
critical_points = sp.solve(A_prime, x)

# Ensure we only keep real, positive solutions
real_critical_points = []
for p in critical_points:
    p_eval = sp.simplify(p)
    if p_eval.is_real and p_eval > 0:
        real_critical_points.append(p_eval)

# Display the results
print("First derivative, A'(x):", A_prime)
print("Critical points (possible min avg cost):", real_critical_points)

# ---------------------------------------------------------------
# Test the critical point using the second derivative
# ---------------------------------------------------------------
A_double_prime = sp.diff(A_prime, x)  # Second derivative

for point in real_critical_points:
    second_derivative_value = A_double_prime.subs(x, point)
    if second_derivative_value > 0:
        nature = "Local Minimum (Optimal Production Level)"
    elif second_derivative_value < 0:
        nature = "Local Maximum"
    else:
        nature = "Inconclusive"
    print(f"At x = {point}, A''(x) = {second_derivative_value}, Nature: {nature}")

# ---------------------------------------------------------------
# Compute the minimum average cost per unit
# ---------------------------------------------------------------
if real_critical_points:
    optimal_x = real_critical_points[0]  # Assuming we found a valid minimum
    min_avg_cost = A.subs(x, optimal_x)
    print(f"The minimum average cost per unit is: {min_avg_cost} at x = {optimal_x}")
else:
    print("No valid critical points found.")

# ---------------------------------------------------------------
# Plot to visualize the behavior of A(x)
# ---------------------------------------------------------------
x_vals = np.linspace(1, 100, 500)  # Avoid x = 0 (division error)
A_vals = [A.subs(x, val) for val in x_vals]

plt.figure(figsize=(10, 6))
plt.plot(x_vals, A_vals, label="A(x) = C(x)/x (Average Cost per Unit)", color="blue", linewidth=2)

# Highlight minimum average cost point
if real_critical_points:
    plt.scatter(float(optimal_x), float(min_avg_cost), color="red", label=f"Min Avg Cost (x={optimal_x})", zorder=3)

plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Average Cost per Unit vs. Production Level")
plt.xlabel("Units Produced (x)")
plt.ylabel("Average Cost per Unit")
plt.legend()
plt.grid()
plt.show()
```

## Problem 4

The following diagram shows a rectangular area ABCD enclosed on three sides by 60 m of fencing, and on the fourth by a wall AB.

image.jpeg


Find the dimensions of the rectangle that enclose maximum area, and then find the maximum area.

```python
# ---------------------------------------------------------------
# Define variables and the two functions: Perimeter and Area
# ---------------------------------------------------------------
x = sp.Symbol('x')  # Width (perpendicular to the wall)
y = 60 - 2*x  # Express length in terms of x using the constraint equation
A = x * y  # Area function

# ---------------------------------------------------------------
# Find critical points
# ---------------------------------------------------------------
A_prime = sp.diff(A, x)  # First derivative of A(x)

# Solve A'(x) = 0 to find the critical points
critical_x = sp.solve(A_prime, x)
critical_x = [p for p in critical_x if p > 0]  # Only positive x makes sense

# Display the results
print("First derivative, A'(x):", A_prime)
print("Critical points (possible max area):", critical_x)

# ---------------------------------------------------------------
# Test the critical point using the second derivative
# ---------------------------------------------------------------
A_double_prime = sp.diff(A_prime, x)  # Second derivative

for point in critical_x:
    second_derivative_value = A_double_prime.subs(x, point)
    if second_derivative_value < 0:
        nature = "Local Maximum (Optimal Width)"
    elif second_derivative_value > 0:
        nature = "Local Minimum"
    else:
        nature = "Inconclusive"
    print(f"At x = {point}, A''(x) = {second_derivative_value}, Nature: {nature}")

# ---------------------------------------------------------------
# Compute the maximum area and corresponding dimensions
# ---------------------------------------------------------------
if critical_x:
    optimal_x = critical_x[0]  # The best width
    optimal_y = 60 - 2*optimal_x  # Compute corresponding length
    max_area = A.subs(x, optimal_x)  # Maximum area

    print(f"The optimal width is: {optimal_x} meters")
    print(f"The optimal length is: {optimal_y} meters")
    print(f"The maximum enclosed area is: {max_area} square meters")
else:
    print("No valid critical points found.")

# ---------------------------------------------------------------
# Plot to visualize the behavior of A(x)
# ---------------------------------------------------------------
x_vals = np.linspace(0, 30, 500)  # x should be between 0 and 30
A_vals = [A.subs(x, val) for val in x_vals]  # Compute A(x)

plt.figure(figsize=(10, 6))
plt.plot(x_vals, A_vals, label="A(x) = x(60 - 2x) (Area)", color="blue", linewidth=2)

# Highlight maximum area point
if critical_x:
    plt.scatter(float(optimal_x), float(max_area), color="red", label=f"Max Area (x={optimal_x})", zorder=3)

plt.axhline(0, color="black", linewidth=0.8, linestyle="--")
plt.axvline(0, color="black", linewidth=0.8, linestyle="--")
plt.title("Maximum Enclosed Area vs. Width")
plt.xlabel("Width (x)")
plt.ylabel("Enclosed Area A(x)")
plt.legend()
plt.grid()
plt.show()
```

```python
# ---------------------------------------------------------------
# Define variables and functions
# ---------------------------------------------------------------
x = sp.Symbol('x', real=True, positive=True)  # x represents how far upstream the crocodile swims
T = 7 * sp.sqrt(36 + x**2) + 4 * (20 - x)  # Time function to minimize

# First derivative (gradient)
T_prime = sp.diff(T, x)

# ---------------------------------------------------------------
# Lambdify the functions
# ---------------------------------------------------------------
T_func = sp.lambdify(x, T, 'numpy')  # Converts T(x) into a numerical function
T_prime_func = sp.lambdify(x, T_prime, 'numpy')  # Converts T'(x) into a numerical function

# ---------------------------------------------------------------
# Calculate the time taken if the crocodile does not travel on land
# ---------------------------------------------------------------
x_no_land = 0  # The crocodile swims directly across the river
T_no_land = T_func(x_no_land)
print(f"Time taken if the crocodile does NOT travel on land: {T_no_land:.2f} tenths of a second")

# ---------------------------------------------------------------
# Calculate the time taken if the crocodile swims the shortest distance possible
# ---------------------------------------------------------------
x_shortest_distance = 20  # The crocodile swims directly to the prey
T_shortest_distance = T_func(x_shortest_distance)
print(f"Time taken if the crocodile swims the shortest distance possible: {T_shortest_distance:.2f} tenths of a second")

# ---------------------------------------------------------------
# Finding the minimum time using Gradient Descent
# ---------------------------------------------------------------
# Gradient Descent Hyperparameters
alpha = 0.01  # Learning rate
xk = 10  # Initial guess (starting value for x)
tol = 1e-6  # Tolerance for stopping criterion
max_iter = 1000  # Maximum iterations
k = 0  # Iteration counter

# Gradient Descent Algorithm
history_x = [xk]  # Store x values for visualization
history_T = [T_func(xk)]  # Store corresponding T values

while k < max_iter:
    x_new = xk - alpha * T_prime_func(xk)  # Gradient Descent Update Rule
    history_x.append(x_new)
    history_T.append(T_func(x_new))

    # Check for convergence
    if abs(x_new - xk) < tol:
        break

    xk = x_new
    k += 1

# Optimal results
optimal_x = xk
optimal_time = T_func(optimal_x)

print(f"Optimal x (distance upstream): {optimal_x:.2f} meters")
print(f"Minimum possible time: {optimal_time:.2f} tenths of a second")

# ---------------------------------------------------------------
# Plot the Time Function and Gradient Descent Path
# ---------------------------------------------------------------
x_vals = np.linspace(0, 20, 400)  # x values for plotting
T_vals = T_func(x_vals)  # Compute T(x) values

plt.figure(figsize=(10, 6))
plt.plot(x_vals, T_vals, label="Time Function T(x)", color="blue", linewidth=2)
plt.scatter(history_x, history_T, color="red", label="Gradient Descent Steps", zorder=3)
plt.scatter(optimal_x, optimal_time, color="green", label=f"Min Time (x={optimal_x:.2f})", s=100, edgecolor='black')

plt.xlabel("x (meters upstream)")
plt.ylabel("Time T(x) (tenths of a second)")
plt.title("Minimizing the Crocodile's Travel Time")
plt.legend()
plt.grid()
plt.show()


```
